#### Search for an address
<https://docs.microsoft.com/en-us/azure/azure-maps/how-to-search-for-address>


This query retrieves a generic search. To get more accurate results, the query should be more precise as well.

```python
address = 'Avenida Navarra'

x = ('https://atlas.microsoft.com/search/address/json?&subscription-key={}&api-version=1.0&language=en-US&query={}').format(subscriptionKey, address)

session = aiohttp.ClientSession()

result = await (await session.get(x)).json()
result
```

#### Fuzzy search
https://docs.microsoft.com/en-us/azure/azure-maps/how-to-search-for-address

This query retrieves what can be found in the nearby following a description.

##### Use cases:
1. Following a credit card payment in a shop, the bank could get instantly what is nearby the costumer and send accurate alocated offers.

```python
fuzzy = 'cajero'
lat = '42.816666'
lon = '-1.650000'
radius = '400'
limit = '5'

x = ('https://atlas.microsoft.com/search/fuzzy/json?&api-version=1.0&subscription-key={}&language=en-US&query={}&lat={}&lon={}&radius={}&limit={}').format(subscriptionKey, fuzzy, lat, lon, radius, limit)

session = aiohttp.ClientSession()

await (await session.get(x)).json()
```
