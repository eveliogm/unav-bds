```python
url = 'https://servicios.ine.es/wstempus/js/ES/SERIES_TABLA/4247?page=1'

r = requests.get(url)
data = r.json()
#prettify
pd.DataFrame(data).head(5)
```

```python
url = 'https://servicios.ine.es/wstempus/js/ES/SERIES_TABLA/4247?page=1'

r = requests.get(url)
r.content
```

```python
r.json()
```

```python
# dynamic query
url = 'https://servicios.ine.es/wstempus/js/ES/SERIES_TABLA/{code}?page=1'

code = '4247'
url = url.format(code=code)

r = requests.get(url)
data = r.json()
pd.DataFrame(data).head(5)
```
