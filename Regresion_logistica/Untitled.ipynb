```python
import requests
from lxml.html import fromstring
import pandas as pd
```

```python
response = requests.get('https://free-proxy-list.net/')
parser = fromstring(response.text)
parser.xpath('/*')
```

```python
table = parser.xpath('//*[@id="list"]/div/div[2]/div/table')
```

```python
rows = parser.xpath('//*[@id="list"]/div/div[2]/div/table//tr')
```

```python
columns = rows[0].xpath('.//text()')
```

```python
data = []
for row in rows[1:]:
    data.append(row.xpath('.//text()'))
```

```python
pd.DataFrame(data=data, columns=columns)
```
