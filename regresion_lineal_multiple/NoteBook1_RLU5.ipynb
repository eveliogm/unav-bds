# Introducción


---



Modelo de regresión lineal univarible 


El objetivo de la regresión lineal es encontrar la recta que mejor se ajuste a los datos observados. Posteriormente, esa recta (modelo) la utilizaremos para predecir valores para nuevos datos.

En el caso más simple (con una variable), este modelo no es más que la recta: $y = ax + b$. 

Es decir, el modelo representa la forma en la que la variable de salida $y$ se obtiene a partir de las variables de entrada $x$. La forma del modelo de la regresión lineal utiliza para ello los parámetros $a$ y $b$. Sin embargo, en aprendizaje automático es más habitual denotar estos parámetros como $\theta_1$ y $\theta_0$, respectivamente.

$$ y = h_\theta(x) = \theta_0 + \theta_1 x $$

Para el problema en cuestión

$$ precio = h_\theta(superficie) = \theta_0 + \theta_1 superficie $$

Una vez establecido el modelo, el problema de aprendizaje se reduce a aprender los parámetros del modelo utilizando los datos disponibles.



Definido el modelo, necesitamos un algoritmo de aprendizaje que nos permita obtener los parámetros del modelo final.


Antes de ver cómo aprender los parámetros, vamos a centrarnos en enterder qué representan.

A continuación, al tratarse de un problema de regresión multivariable (con más de una variable predictoras) necesitamos eliminar el efecto de escala asociado a las distintas unidades de medida de cada una de las variables. 

Con este objetivo, y en la misma línea de la estandarización que vimos en el módulo 4 donde asumíamos que la distribución de frecuencias de nuestras variables seguía una distribución normal (o gaussiana).

Para la estandarización calculábamos los nuevos valores $X^*$ a partir de:

$X^* = \dfrac{X-mean(X)}{SD(X)}$

donde SD(X) representa la desviación estándar y mean(X) es el valor medio o esperanza.


En esta [página](https://scikit-learn.org/stable/auto_examples/preprocessing/plot_all_scaling.html#robustscaler) de la librería sklearn podemos ver la comparación entre distintas funciones para estandarizar, normalizar o realizar una transformación Box-Cox de los datos.

## Construcción del modelo<a name="modelo"></a> 






---

En nuestro modelo de regresión lineal tenemos que la variable dependiente $Y$ (precio de la vivienda) es una función lineal de la variable independiente $X$  (superficie) de tal forma que la expresión:

 $Y = a * X + b$

 donde $a$ y $b$ son los coeficiente o parámetros del ajuste lineal. $a$ es la pendiente de la recta y $b$ es el punto de corte de la recta en el eje de ordenadas (en inglés *intercept*).

También se puede escribir como:

 $Y = a * \text{superficie}  + b$


---

- El valor de la pendiente $a$ lo determinamos mediante la función **lr.coef_**
- El término independiente $b$ lo determinamos mediante la función **lr.intercept_**





```python
from sklearn.linear_model import LinearRegression #Importamos la función regresión lineal
import sklearn.metrics as metrics
from sklearn.metrics import mean_squared_error #Importamos la función MSE
#Obtención del modelo de regresión lineal para la muestra de entrenamiento
lr= LinearRegression() 
lr.fit(X_train, Y_train) #construcción del modelo a partir de los datos de entrenamiento.
#Cálculo de los valores predichos a partir de la muestra de entrenamiento
Y_train_pred= lr.predict(X_train)
#Cálculo de los valores predichos a partir de la muestra de test
Y_test_pred= lr.predict(X_test)
```

### Cómo afectan los parámetros al modelo

Dados los parámetros del modelo de regresión lineal univarible:
$$\theta = \begin{pmatrix} \theta_0 \\ \theta_1 \end{pmatrix} $$

Vamos a definir primero una función que modele la regresión lineal basada en los parámetros $\theta$.

```python
def linear_regression_model(thetas, x=np.linspace(0,4)):
    y = thetas[0, 0] + thetas[1, 0] * x
    return y
```

Vamos a representar gráficamente nuestro modelo lineal para distintos valores de $\theta_0$ y $\theta_1$ de tal forma que podamos ver cómo afectan los diferentes parámetros al modelo.


```python
plt.figure(figsize=(18, 5.5))
x_pp = np.linspace(0, 4)
# Definimos los pares de valores de theta a probar
thetas = np.array([[1.5, 0],
                 [0, 0.5],
                 [1, 0.5]])
# Dibujamos las gráficas en una figura
fig, ax = plt.subplots(1,thetas.shape[0], figsize=(18, 6), sharex=True, sharey=True)
for i in range(thetas.shape[0]):
    thetas_i = thetas[i, :].reshape(-1,1)
    ax[i].plot(x_pp, linear_regression_model(thetas_i, x_pp), linewidth=3, color='red')
    ax[i].set_xlabel('X')
    ax[i].set_ylabel('Y')
    ax[i].set_xlim([0, 4])
    ax[i].set_ylim([0, 3])
    ax[i].annotate(r'$h_\theta(x)={} + {}x$'.format(thetas_i[0, 0], thetas_i[1, 0]), (0.5, 2.5), fontsize=16)
    ax[i].annotate(r'$\theta_0 = {}, \theta_1 = {}$'.format(thetas_i[0, 0], thetas_i[1, 0]), (0.5, 2.3), fontsize=16)
plt.show()
```

## Evaluación de la bondad del ajuste
  
Las métricas son medidas de la bondad del ajuste, es decir, nos dicen cuán bien el modelo de regresión lineal se ajusta a las observaciones.

$\bullet$ MAE es el **error absoluto medio** (*Mean Absolute Error*) y se calcula como el promedio del valor absoluto de la diferencia entre los valores predichos y los valores observados.

$$MAE = \dfrac{1}{n}\sum\limits_{i=1}^n |Y(i)-Y_{pred}(i))|$$

$\bullet$ MSE es el **error cuadrático medio** (*Mean Square Error*) que se calcula a partir de los errores de aproximación o residuos como:


$$MSE = \dfrac{1}{n}\sum\limits_{i=1}^n (Y(i)-Y_{pred}(i)) ^2 $$


$\bullet$ RMSE es el **error estándar de la estimación** (*Root Mean Square Error*). Es una de las métricas más importantes en regresión y se calcula tomando la raíz cuadrada del error cuadrático medio por lo que recupera las mismas dimensiones que la variable objetivo o respuesta: 

$RMSE = \sqrt{MSE}$

El error estándar de la estimación es una medida de la precisión de los estimadores ($\theta_0$, $\theta_1$) obtenidos en la regresión.

$\bullet$ $R^2$ es el **coeficiente de determinación** que nos indica la proporción de la información de los datos que se explican por el modelo:


$$R^2= 1 - \dfrac{RSS}{TSS} \; \; ; \; \; \text{con} \; \; R^2 \in [0,1] $$

Donde RSS se corresponde con la suma del cuadrado de los residuos (*Residual Sum of Squares*):

$$RSS = \sum\limits_{i=1}^n (Y(i)-Y_{pred}(i))^2$$

y TSS representa la variabilidad total en $R^2$ o varianza de la variable respuesta (*Total Sum of Squares*):

$$TSS = \sum\limits_{i=1}^n (Y(i)-\overline{Y})^2$$

$\overline{Y}$ representa el valor medio de Y.



---

* A la métrica RSS (*Residual Sum of Squares*) también la encontraremos con las siglas SSE ({*Sum of Squares Error*).
* A la métrica TSS (*Total Sum of Squares*) también la encontraremos con las siglas SST (*Sum of Squares Total*).



Definimos la función *calculametricas* con las métricas de la regresión:

```python
def calculametricas(y_test, y_pred):
    mae = metrics.mean_absolute_error(y_test, y_pred) #Error absoluto medio (MAE)
    mse = metrics.mean_squared_error(y_test, y_pred) #Error cuadrático medio (MSE)
    rmse = math.sqrt(mse) #Error estándar de la estimación (RMSE)
    r2 = metrics.r2_score(y_test, y_pred) #Cálculo del coeficiente de determinación R2
    print("MAE : ",mae)
    print("MSE : ",mse)
    print("RMSE : ",rmse)
    print("R2 : ",r2)
    return mae,mse,rmse,r2
```

>> Este algoritmo incluye tres tipos de funciones de error (de coste o *loss*): ‘squared_error’, ‘huber’, ‘epsilon_insensitive’ y ‘squared_epsilon_insensitive’.
>> Además de tres técnicas de regularización: 'l1', 'l2' y 'elasticnet'.


---

**SGDRegressor: Regresión Lineal con Descenso por Gradiente Incremental**

El gradiente de la función de error se estima para cada observación a la vez que los coeficientes de aprendizaje del modelo se actualizan.

En esta función tenemos los argumentos más destacables son:
* `max_iter`: número de iteraciones del descenso por gradiente
* `learning_rate`: establecer a `constant` para simular los mismos efectos.
* `eta0`: este parámetro es el factor de aprendizaje, el que nosotros hemos visto como $\alpha$ que suele tomar valores en el intervalo $[10^{-1}, 10^{-5}]$ siendo `eta0` = 0.01 el valor por defecto. No confundir con el parámetro `alpha` que también aparece en este método y se refiere al proceso de regularización.




## **Método de Descenso por Gradiente**

El algoritmo del descenso por gradiente es uno de los más  importantes en *Machine Learning* debido a que es utilizado para entrenar multitud de modelos, entre ellos las redes neuronales. En realidad, es un algoritmo de minimización muy sencillo que funciona de la siguiente forma:

1. Comenzamos con valores aleatorios de $\theta$.
2. Cambiamos los valores de $\theta$ con el objetivo de reducir el error $J(\theta)$ moviéndonos en la dirección contraria al gradiente (hasta que alcanzamos el mínimo).

El **descenso por gradiente incremental** representa un extremo del descenso por gradiente por "mini-lotes" donde cada lote está formado por un único registro u observación. Con cada registro, calculamos los gradiente y actualizamos los parámetros. De nuevo, suele ser más rápido que la versión por lotes, pero no asegura converger al mínimo global. 

En más detalle, los pasos son los siguientes:
1. Asignamos a $\theta=\{\theta_0, \ldots, \theta_n \}$ valores aleatorios
2. Repetimos hasta convergencia (nº de iteraciones o $|J(\theta_{it}) - J(\theta_{it-1})| < \mu$ )

$$ \theta_j = \theta_j - \alpha \frac{\partial J(\theta)}{\partial\theta_j} \text{  para } j=0,\ldots, n $$



donde,

* Los $\theta_j$ se actualizan simultáneamente.
* $\alpha$ es el  *factor de aprendizaje* que mide cuánto modificamos el valor del parámetro.
* $\frac{\partial J(\theta)}{\partial\theta_j}$ es la *pendiente de la recta tangente* al punto actual ($\theta$).
* Restamos un valor proporcional a la derivada parcial porque queremos ir en la dirección contraria a la indica la pendiente.

Las derivadas parciales se calculan a partir de la función de error:
$$ J(\theta) = \frac{1}{2m}\sum^m_{i=1}(h_\theta(x^{(i)})-y^{(i)})^2, \ \ \ \ h_{\theta}(x) = \theta^Tx  $$

De donde  la actualización del descenso por gradiente tiene la expresión:

$$ \theta_j = \theta_j - \alpha \frac{1}{m}\sum_{i=1}^m (h_\theta(x^{(i)}) - y^{(i)})\cdot x_j^{(i)}   \text{ para  } j=0,\ldots,n   $$


```python
from sklearn.linear_model import SGDRegressor
#Obtención del modelo de regresión lineal con descenso por gradiente: 
#Ensayar eta0=0.01, 0.001, 0.0001, 00001; default=0.01
#Ensayar tol=0.01-0.0001; default=0.001
#loss='squared_error'
sgd = SGDRegressor(loss='squared_error',learning_rate='constant', max_iter=2000, eta0=0.001, tol=0.0001) # eta0 es el parámetro alpha de la teoría de descenso por gradiente
#Entrenamos el modelo con la muestra de entrenamiento
sgd.fit(X_train, Y_train) #construcción del modelo a partir de los datos de entrenamiento.
#Cálculo de los valores predichos a partir de la muestra de entrenamiento
Y_train_pred_sgd= sgd.predict(X_train)
#Cálculo de los valores predichos a partir de la muestra de test
Y_test_pred_sgd= sgd.predict(X_test)
```
