
# Base de datos



---


 Generación de datos sintéticos para la variable respuesta **y** que se corresponde con una función polinómica de segundo grado ("parábola") y la variable **x**.



```python
def gen_non_linear(n, seed=None):
    if seed is not None:
        np.random.seed(seed)
    # Generamos un conjunto de observaciones (x, y) 
    x = np.random.rand(n).reshape(-1, 1) #generación aleatoria
    y = 1 - x + x*x*2  - x*x*x  + np.random.rand(n).reshape(-1, 1)*0.1 #definición de de la función polinómica
    y = y.reshape(-1, 1)

    df = pd.DataFrame(np.hstack((x, y)), columns=['x', 'y'])
    df.head(5)
    
    return x, y, df

x, y, df = gen_non_linear(500, seed=3320)
df.head()
```

Visualizamos los datos para comprobar el comportamiento no-lineal de una función cuadrática (polinomio de grado 2).

```python
plt.figure(figsize=(8, 6))
plt.plot(x, y, '.', markersize=12, color='blue')
plt.xlabel('x')
plt.ylabel('y')
```

# Construcción del modelo polinomial

Nuestro modelo ahora tiene esta forma
$$ h_\theta(x) = \theta_0 + \theta_1 x + \theta_2 x^2 $$
 

1. Como hemos añadido una nueva variable, tenemos un nuevo parámetro. No obstante, realmente nuestro problema tiene una única variable (x), y el $\theta_2$ en este caso corresponde al término $x^2$. 
>Por esta razón, podemos seguir dibujando la regresión, aunque veremos que ahora deja de ser una recta (a pesar de que en las tres dimensiones sí seguirá siendo un híper-plano).

2. Al introducir variables polinomiales generamos rangos muy diferentes entre las variables. Por tanto, es totalmente indispensable estandarizar las variables.

```python
from sklearn.preprocessing import StandardScaler #importamos la librería para estandarizar
import sklearn.metrics as metrics
```

## Creamos la matriz de variables predictoras $(x_o, x_1, x_2)$


---


**N.B.:** De acuerdo con la función $y$ que hemos definido al inicio (en la generación sintética de los datos) 

>> $x_0=1, \; x_1= x,\;  x_2=x^2$

Estas son nuestras variables predictoras a las que nos referimos como *variables polinómicas* ("polynomial features").

```python
# Creamos las variables polinomiales
poly_features = PolynomialFeatures(degree=3, include_bias=False) #include_bias incluye la columna de 1s si seleccionásemos True (default value)
poly_features.fit(x)
x_poly = poly_features.transform(x)
dfx_poly=pd.DataFrame(x_poly)
dfx_poly = dfx_poly.rename(columns = {0:'x_0',1:'x_1',2:'x_2'})
dfx_poly.head()
```

## Representación en el espacio  $(x, x^2, y)$

Comprobamos que la dispersión de puntos se encuentra situada sobre un plano. Es decir, la recta de la regresión de un problema de 2 variables $(x,y)$ se convierte en el plano de regresión para un problema de 3 variables $(x, x^2, y)$.


```python
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
ax.xaxis.label.set_size(20)
ax.yaxis.label.set_size(20)
ax.zaxis.label.set_size(20)
# Representamos los datos de entrada
ax.scatter(x, x**2, y, marker='.', color='blue')
ax.set_xlabel('$x (= x_1)$')
ax.set_ylabel(r'$x^2 (=x_2)$')
ax.set_zlabel('y')
```

## Evaluación de la bondad del ajuste
  
Las métricas son medidas de la bondad del ajuste, es decir, nos dicen cuán bien el modelo de regresión lineal se ajusta a las observaciones.

$\bullet$ MAE es el **error absoluto medio** (*Mean Absolute Error*) y se calcula como el promedio del valor absoluto de la diferencia entre los valores predichos y los valores observados.

$$MAE = \dfrac{1}{n}\sum\limits_{i=1}^n |Y(i)-Y_{pred}(i))|$$

$\bullet$ MSE es el **error cuadrático medio** (*Mean Square Error*) que se calcula a partir de los errores de aproximación o residuos como:


$$MSE = \dfrac{1}{n}\sum\limits_{i=1}^n (Y(i)-Y_{pred}(i)) ^2 $$


$\bullet$ RMSE es el **error estándar de la estimación** (*Root Mean Square Error*). Es una de las métricas más importantes en regresión y se calcula tomando la raíz cuadrada del error cuadrático medio por lo que recupera las mismas dimensiones que la variable objetivo o respuesta: 

$RMSE = \sqrt{MSE}$

El error estándar de la estimación es una medida de la precisión de los estimadores ($\theta_0$, $\theta_1$) obtenidos en la regresión.

$\bullet$ $R^2$ es el **coeficiente de determinación** que nos indica la proporción de la información de los datos que se explican por el modelo:


$$R^2= 1 - \dfrac{RSS}{TSS} \; \; ; \; \; \text{con} \; \; R^2 \in [0,1] $$

Donde RSS se corresponde con la suma del cuadrado de los residuos (*Residual Sum of Squares*):

$$RSS = \sum\limits_{i=1}^n (Y(i)-Y_{pred}(i))^2$$

y TSS representa la variabilidad total en $R^2$ o varianza de la variable respuesta (*Total Sum of Squares*):

$$TSS = \sum\limits_{i=1}^n (Y(i)-\overline{Y})^2$$

$\overline{Y}$ representa el valor medio de Y.



---

* A la métrica RSS (*Residual Sum of Squares*) también la encontraremos con las siglas SSE ({*Sum of Squares Error*).
* A la métrica TSS (*Total Sum of Squares*) también la encontraremos con las siglas SST (*Sum of Squares Total*).



Definimos la función *calculametricas* con las métricas de la regresión:

```python
def calculametricas(y_test, y_pred):
    mae = metrics.mean_absolute_error(y_test, y_pred) #Error absoluto medio (MAE)
    mse = metrics.mean_squared_error(y_test, y_pred) #Error cuadrático medio (MSE)
    rmse = math.sqrt(mse) #Error estándar de la estimación (RMSE)
    r2 = metrics.r2_score(y_test, y_pred) #Cálculo del coeficiente de determinación R2
    print("MAE : ",mae)
    print("MSE : ",mse)
    print("RMSE : ",rmse)
    print("R2 : ",r2)
    return mae,mse,rmse,r2
```

# Ejemplo de sobreajuste con [sklearn](https://scikit-learn.org/stable/auto_examples/model_selection/plot_underfitting_overfitting.html)




---



Aplicaremos la regresión lineal con variables polinomiales para construir tres modelos diferentes que ajusten comportamientos no-lineales:

* Una recta: $P_1(x)$

* Un polinomio de grado 4: $P_4(x)$

* Un polinomio de grado 15: $P_{15}(x)$



---

**N.B.:** Hablamos de **sobreajuste de un modelo** cuando el modelo de regresión se ajusta perfectamente a los datos sin aportar ninguna información explicativa sobre comportamiento de los datos (la relación entre las variables) ni tampoco predecir correctamente el comportamiento de nuevos datos sobre el mismo problema.
---

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score


def true_fun(X):
    return np.cos(1.5 * np.pi * X)


np.random.seed(0)

n_samples = 30
degrees = [1, 2,3,6]

X = np.sort(np.random.rand(n_samples))
y = true_fun(X) + np.random.randn(n_samples) * 0.1

plt.figure(figsize=(18, 8))
for i in range(len(degrees)):
    ax = plt.subplot(1, len(degrees), i + 1)
    plt.setp(ax, xticks=(), yticks=())

    polynomial_features = PolynomialFeatures(degree=degrees[i], include_bias=False)
    linear_regression = LinearRegression()
    pipeline = Pipeline(
        [
            ("polynomial_features", polynomial_features),
            ("linear_regression", linear_regression),
        ]
    )
    pipeline.fit(X[:, np.newaxis], y)

    # Evaluate the models using crossvalidation
    scores = cross_val_score(
        pipeline, X[:, np.newaxis], y, scoring="neg_mean_squared_error", cv=10
    )

    X_test = np.linspace(0, 1, 100)
    plt.plot(X_test, pipeline.predict(X_test[:, np.newaxis]), label="Modelo de regresión lineal")
    plt.plot(X_test, true_fun(X_test), label="Función real")
    plt.scatter(X, y, edgecolor="b", s=20, label="Datos de entrada")
    plt.xlabel("x")
    plt.ylabel("y")
    plt.xlim((0, 1))
    plt.ylim((-2, 2))
    plt.legend(loc="best")
    plt.title(
        "Grado del polinomio {}\nMSE = {:.2e}(+/- {:.2e})".format(
            degrees[i], -scores.mean(), scores.std()
        )
    )
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score


def true_fun(X):
    return np.cos(1.5 * np.pi * X)


np.random.seed(0)

n_samples = 30
degrees = range(1,100,1)

X = np.sort(np.random.rand(n_samples))
y = true_fun(X) + np.random.randn(n_samples) * 0.1

scores_all = []
for i in range(len(degrees)):

    polynomial_features = PolynomialFeatures(degree=degrees[i], include_bias=False)
    linear_regression = LinearRegression()
    pipeline = Pipeline(
        [
            ("polynomial_features", polynomial_features),
            ("linear_regression", linear_regression),
        ]
    )
    pipeline.fit(X[:, np.newaxis], y)

    # Evaluate the models using crossvalidation
    scores = cross_val_score(
        pipeline, X[:, np.newaxis], y, scoring="neg_mean_squared_error", cv=10
    )
    scores_all.append(-scores.mean())
```

```python
plt.figure(figsize=(10, 5))
plt.plot(degrees,scores_all)
plt.xlabel("degree")
plt.ylabel("MSE")
```
